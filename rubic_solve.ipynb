```python
def same_face_spin(rubik,topleft):
    #samefacespin
    #mid
    #304 -> 405
    temp = rubik[topleft+2+12]
    rubik[topleft+2+12] = rubik [topleft+1]
    #405 -> 504
    temp2 = rubik[topleft+1+12*2]
    rubik[topleft+1+12*2] = temp
    #504 -> 403
    temp = rubik[topleft + 12]
    rubik[topleft+12] = temp2
    #403 -> 304
    rubik[topleft+1] = temp
    #edge
    #303 -> 305
    temp = rubik[topleft+2]
    rubik[topleft+2] = rubik[topleft]
    #305 -> 505
    temp2 = rubik[topleft+2+12*2]
    rubik[topleft+2+12*2] = temp
    #505 -> 503
    temp = rubik[topleft+12*2]
    rubik[topleft+12*2] = temp2
    #503 -> 303
    rubik[topleft] = temp
    return rubik

def resame_face_spin(rubik,topleft):

    temp = rubik[topleft+12]
    rubik[topleft+12] = rubik[topleft+1]

    temp2 = rubik[topleft+12*2+1]
    rubik[topleft+12*2+1] = temp

    temp = rubik[topleft+12+2]
    rubik[topleft+12+2] = temp2

    rubik[topleft+1] = temp

    temp = rubik[topleft+12*2]
    rubik[topleft+12*2] = rubik[topleft]

    temp2 = rubik[topleft+12*2+2]
    rubik[topleft+12*2+2] = temp

    temp = rubik[topleft+2]
    rubik[topleft+2] = temp2

    rubik[topleft] = temp

    return rubik
```

```python
class Goal(Exception):
    pass
```

```python
class Node:
    def __init__(self, state, result) -> None:
        self.state = state
        self.branching_factor = {
            'f':None,
            'r':None,
            'u':None,
            'b':None,
            'l':None,
            'd':None,
            'f_re':None,
            'r_re':None,
            'u_re':None,
            'b_re':None,
            'l_re':None,
            'd_re':None,
        }
        self.result = result
```

```python
class BFS_TEST_CASE: 
    def __init__(self) -> None:
        self.root = None
        self.frontier = []
        self.action_lists = list(Node(None, None).branching_factor.keys())

    def set_start_state(self, state):
        self.root = Node(state, [])
        self.frontier.append(self.root)

    def next_state(self, act, node):
        node.branching_factor[act] = Node(action(act, node.state.copy()), node.result + [act])
        self.frontier.append(node.branching_factor[act])
        # if check_goal(node.branching_factor[act].state):
        #     raise Goal
    
    def expand_state(self, node):
        if not self.root:
            self.root = node
            self.frontier.append(node)
            return
        for action in self.action_lists:
            self.next_state(action, node)
    
    def frontier_state(self):
        if not len(self.frontier):
            return
        queue = self.frontier.copy()
        for node in queue:
            self.expand_state(node)
            self.frontier.pop(0)
```

```python
class BFS: 
    def __init__(self) -> None:
        self.root = None
        self.frontier = []
        self.action_lists = list(Node(None, None).branching_factor.keys())

    def set_start_state(self, state):
        self.root = Node(state, [])
        self.frontier.append(self.root)

    def next_state(self, act, node):
        node.branching_factor[act] = Node(action(act, node.state.copy()), node.result + [act])
        self.frontier.append(node.branching_factor[act])
        if check_goal(node.branching_factor[act].state):
            raise Goal
    
    def expand_state(self, node):
        if not self.root:
            self.root = node
            self.frontier.append(node)
            return
        for action in self.action_lists:
            self.next_state(action, node)
    
    def frontier_state(self):
        if not len(self.frontier):
            return
        queue = self.frontier.copy()
        for node in queue:
            self.expand_state(node)
            self.frontier.pop(0)
    
    def solve(self):
        if not self.root:
            return
        print('start state:')
        show_rubik(self.root.state)
        print()
        print('result action :', self.frontier[-1].result)
        print()
        state = self.root.state.copy()
        i = 1
        for act in self.frontier[-1].result:
            print('action ', i, ':', act)
            state = action(act, state)
            show_rubik(state)
            print()
            i+=1

```

```python
class BDS():
    def __init__(self) -> None:
        self.start_tree = BFS_TEST_CASE()
        self.goal_tree = BFS_TEST_CASE()

        self.goal_tree.set_start_state(gen_rubik().copy())

        self.intersect_state = []
        self.actions = []
    
    def set_start_state(self, state):
        self.start_tree.set_start_state(state)
    
    def simultaneously_searches(self):
        self.start_tree.frontier_state()
        self.intersection()
        self.goal_tree.frontier_state()
        self.intersection()
    
    def intersection(self):
        intersect_state = [i.state for i in self.start_tree.frontier if i.state in [j.state for j in self.goal_tree.frontier]]
        if len(intersect_state):
            self.intersect_state = intersect_state
            raise Goal 
    
    def result_actions(self):
        start_result_action = [i.result for i in self.start_tree.frontier if i.state == self.intersect_state[0]][0]
        goal_result_action = [i.result for i in self.goal_tree.frontier if i.state == self.intersect_state[0]][0]
        print(start_result_action,goal_result_action)
        reverse = []
        for action in goal_result_action[::-1]:
            if '_re' in action:
                reverse.append(action[0])
            else:
                reverse.append(action + '_re')
        return start_result_action + reverse
    
    def solve(self):
        if not self.start_tree.root:
            return
        print('start state:')
        show_rubik(self.start_tree.root.state)
        print()
        self.actions = self.result_actions()
        print('result action :', self.actions)
        print()
        state = self.start_tree.root.state.copy()
        i = 1
        for act in self.actions:
            print('action ', i, ':', act)
            state = action(act, state)
            show_rubik(state)
            print()
            i+=1
```
